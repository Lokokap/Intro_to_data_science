We are given a Markov chain with 4 states:

- Downtown (D)  
- Suburbs (S)  
- Countryside (C)  
- Workshop (W)  

The transition matrix \(P\) is:

\[
P = \begin{bmatrix}
0.3 & 0.7 & 0 & 0 \\
0.2 & 0.5 & 0.3 & 0 \\
0 & 0 & 0.5 & 0.5 \\
0 & 0 & 0 & 1
\end{bmatrix}
\]

### Part 1: Probability of being in Countryside after 10 steps from Downtown

We want:

\[
P(X_{10} = C \mid X_0 = D)
\]

- This is obtained by raising the transition matrix to the 10th power and taking the (D, C) entry:

```python

### Part 5: Expected number of steps to reach Workshop starting from Downtown

Let \(h_D, h_S, h_C, h_W\) denote the expected number of steps to reach the Workshop (W) from each state.  

The equations are:

\[
\begin{cases}
h_W = 0 \\
h_D = 1 + 0.3 h_D + 0.7 h_S \\
h_S = 1 + 0.2 h_D + 0.5 h_S + 0.3 h_C \\
h_C = 1 + 0.5 h_C + 0.5 h_W
\end{cases}
\]

**Step-by-step solution:**

1. \(h_W = 0\)  

2. Solve for \(h_C\):  
\[
h_C = 1 + 0.5 h_C \implies 0.5 h_C = 1 \implies h_C = 2
\]

3. Solve for \(h_S\):  
\[

\]

4. Solve for \(h_D\):  
\[
h_D = 1 + 0.3 h_D + 0.7 h_S = 1 + 0.3 h_D + 0.7 (3.2 + 0.4 h_D) = 3.24 + 0.58 h_D
\]  
\[
0.42 h_D = 3.24 \implies h_D \approx 7.7
\]

Interpretation:

Confidence band shows uncertainty in the EDF due to finite sample size.

Can be used to test if residuals follow a specific distribution (e.g., normality).

````python
# Part 6
# Put the code for part 6 below this line

import numpy as np

# Residuals
residuals = problem2_y_test - y_pred
n = len(residuals)

# Empirical CDF
sorted_residuals = np.sort(residuals)
edf = np.arange(1, n+1) / n

# DKW inequality for 95% confidence
import math
epsilon = math.sqrt(np.log(2/0.05)/(2*n))

# Plot
plt.step(sorted_residuals, edf, where='post', label='EDF of Residuals')
plt.fill_between(sorted_residuals, np.maximum(0, edf - epsilon), np.minimum(1, edf + epsilon),
                 color='lightblue', alpha=0.5, label='95% Confidence Band (DKW)')
plt.xlabel("Residual")
plt.ylabel("ECDF")
plt.title("Empirical Distribution Function of Residuals")
plt.legend()
plt.show()
````

### Part 4: Can we use Gaussian as proposal?

Target PDF:  

\[
f(x) = C e^{-(x^2-2)^2}
\]

- Gaussian is positive everywhere â†’ proposal \(g(x) = \mathcal{N}(0,1)\)
- Problem: target \(f(x)\) has **faster decay than Gaussian**? Actually, \(f(x)\) decays as \(e^{-(x^2-2)^2}\), which is faster than a Gaussian tail.

âœ… It **is possible** to use Gaussian as a proposal.

- We need 
\[
M \ge \sup_x \frac{f(x)}{g(x)}
\] 
to ensure \(f(x) \le M g(x)\). Choose a large enough \(M\) to satisfy this.

**Answer:**

- Yes, Gaussian can work, but the acceptance rate might be low if the Gaussian is not centered near the mode \(x = \sqrt{2}\).  
- We can scale/shift the Gaussian (mean â‰ˆ \(\sqrt{2}\), variance â‰ˆ 1) to improve efficiency.
